```python
# -*- coding: utf-8 -*-
"""
Created on Mon Feb 18 22:06:34 2019

@author: DELL
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data_linear.csv').values
N = data.shape[0]

x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
plt.scatter(x, y)
plt.xlabel('m√©t vu√¥ng')
plt.ylabel('gi√°')
```

```python
# Thi·∫øt l·∫≠p c√°c tham s·ªë 
# Gi√° tr·ªã ƒë·∫ßu c·ªßa tr·ªçng s·ªë
w1 = 0
w0 = 0

# The learning Rate
alpha = 0.0001
# The number of iterations to perform gradient descent
epochs = 50
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of elements in x
n = len(x)

# Define functions for calculating gradients
def dL_w1(w1, w0):
    return 1/n * sum(2 * (w1 * x[k] + w0 - y[k]) * x[k] for k in range(n))

def dL_w0(w1, w0):
    return 1/n * sum(2 * (w1 * x[k] + w0 - y[k]) for k in range(n))
# Perform Gradient Descent
for i in range(epochs):
    w1 = w1 - alpha * dL_w1(w1, w0)
    w0 = w0 - alpha * dL_w0(w1, w0)
# Visualize the data and the fitted line
plt.scatter(x, y)
plt.xlabel('m√©t vu√¥ng')
plt.ylabel('gi√°')
plt.plot((min(x), max(x)), (min(x) * w1 + w0, max(x) * w1 + w0), color='red')  # Fitted line
plt.show()
```
